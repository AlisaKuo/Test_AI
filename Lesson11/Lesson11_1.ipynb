```python
import requests                     #安裝requests套件 (pip 25.0)
from requests import JSONDecodeError

try:
    response = requests.get("http://example.com/api", params={"key":"value"})
    print (response.status_code)    #狀態碼
    print (response.text)           #獲取響應內容(文字)
    print (response.json())         #若回應為JSON格式, 可直接解析為dict
except JSONDecodeError:
    print ("內容不是json格式")
except Exception:
    print ("發生錯誤")
else:
    print ("沒有發生錯誤")   
```

```python
import requests
from requests import HTTPError, JSONDecodeError

try:
    url = 'https://data.moenv.gov.tw/api/v2/aqx_p_432?api_key=af57253c-e838-46da-a1f5-12b43afd75f3&limit=1000&sort=ImportDate desc&format=JSON'
    response = requests.get (url,verify=False)
    response.raise_for_status()
    data_dict:dict = response.json()
    data_list:list = data_dict("fields")
    for item in data_list:
        print(item)
    #print(data_dict)             #str不是資料結構
except HTTPError as e:
    print (f"HTTPError:{e}")
except JSONDecodeError:
    print ("內容不是json格式")
except Exception:
    print ("不知的錯誤")
else:
    print ("下載成功")
```
