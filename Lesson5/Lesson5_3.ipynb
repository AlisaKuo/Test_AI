```python
# 小明要存錢買機車
# 機車每輛30000元
# 他將每月存的錢輸入, 當存款足夠時, 顯示訊息通知

money= 0
time= 0
while money <30000:
    time=time+1
    inputNum = int(input(f"請輸入第{time}個月份的存款金額:"))
    print (f"第{time}個月, {inputNum}元")
    money=money+inputNum

print (f"恭喜! 存了{time}個月, 共 {money}元")
```

```python
money= 0
time= 0
while True:
    time+=1
    inputNum = int(input(f"請輸入第{time}個月份的存款金額:"))
    money=money+inputNum
    print (f"第{time}個月, {inputNum}元")
    if money >=30000:
        break

print (f"恭喜! 存了{time}個月, 共 {money}元")
```

```python
#請設計一個程式
#讓使用者輸入數值, 只加總正偶數值, 不加總正奇整數
#如果輸入負數, 結束程式

sum=0
num=0
while True:
    input_value=int(input(f"請輸入第{num+1}個數值"))
    if input_value <0:
        break
    num+=1
    
    if input_value %2 ==1:
        continue
    sum+= input_value

print(f"加總的次數是{num}, 輸入的正偶數加總是{sum}")

```
