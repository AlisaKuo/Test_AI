```python
#巢狀迴圈

for I in range(1,10):
    print(f"I={I}")
    for J in range(1,10):
        print(f"J={J}",end=" ")
    print()
    print("=="*20)
```

```python
for I in range(1,10):
    for J in range(1,10):
        print(f"{I}*{J}={I*J}",end="\t")
    print()
```

```python
for I in range(1,10):
    for J in range(1,10):
        amount=I*J
        print(str(I)+"*"+str(J)+"="+str(amount),end="\t")
    print()
```
