```python
#巢狀判斷

scores = int(input("請輸入學生分數0~100:"))

if scores >= 90:
    print ("優")   #條件達成時,執行完就跳出if

elif scores >= 60:  #多項選擇
    print ("及格")

else: 
    print ("不及格")
```

```python
#巢狀判斷

scores = int(input("請輸入學生分數0~100:"))

if scores >= 90:
    print ("優")   #條件達成時,執行完就跳出if

elif scores >= 80:  #多項選擇
    print ("甲")

elif scores >=70:
    print ("乙")

elif scores >=60:
    print("丙")

else: 
    print ("丁")
```

```python
#計算BMI
print("====BMI計算====")
height = int(input("請輸入身高 cm:"))
weight = int(input("請輸入體重 kg:"))
BMI = weight/(height/100)**2
print(f"你的BMI是{round(BMI,ndigits=2)}")  #大括弧


if BMI < 18.5:
    print ("體重過輕")

elif BMI <24:
    print ("正常範圍")

elif BMI <27:
    print ("體重過重")

elif BMI <30:
    print ("輕度肥胖")

elif BMI <35:
    print ("中度肥胖")

else:
    print ("重度肥胖")
```
