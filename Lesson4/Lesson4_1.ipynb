```python
# 學生總分為300分
# 有些學生可以加分5%
# 如果加分超過 300, 就以 300 分為準

try:
    scores = int(input("請輸入學生的分數(最高300分):"))

    if scores > 300:
        print("分數不可以超過300")
    else:
        add = input("學生是否符合加分條件(y/n)")
        
        if (add == "y"):   # 單項選擇
            scores *= 1.05
            if scores > 300: # 利用巢狀判斷增加條件
                scores = 300

        print(scores)
        print(int(scores)) # 移除小數點
        print(round(scores)) # 四捨五入
        print(round(scores,ndigits=2)) #顯示到小數點第二位

except ValueError:
    print("請輸入學生的分數, 無法辨識")
except:
    print ("出現錯誤, 請重新執行")

print ("應用程式結束")

```
