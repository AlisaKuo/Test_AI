```python
def sayhello(name):         #自訂func,1個參數,有傳出值
    return f"您好, {name}"

print(sayhello("Alisa"))    #輸入的名字會傳回到 name 參數
```

```python
def get_bmi(w,h):            #2個參數, 傳出一個值
    bmi = w/(h/100) ** 2
    return bmi

def bmi_status(bmi):
    if bmi < 18.5:
        message = "體重過輕"
    elif bmi <24:
        message = "正常範圍"
    elif bmi <27:
        message = "體重過重"
    elif bmi <30:
        message = "輕度肥胖"
    elif bmi <35:
        message = "中度肥胖"
    else:
        message = "重度肥胖"
    
    return message

# 計算 BMI
print("====== BMI計算 ======\n")
h = int(input("請輸入身高 cm:"))
w = int(input("請輸入體重 kg:"))
bmi = get_bmi(w, h)
print(f"您的 BMI 是 {round(bmi,ndigits=2)}")  #大括弧
status = bmi_status(bmi)
print(status)


```
