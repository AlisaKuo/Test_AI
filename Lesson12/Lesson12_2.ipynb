```python
import matplotlib.pyplot as plt

# 1.製作figure
fig = plt.figure()

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
#ax_1.plot(X, Y)

ax_1.plot(X, Y, color="black", marker="s", linestyle="solid",linewidth=1, markersize=8)

# 4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt
# 1.製作figure
fig = plt.figure()      
# 2.1*3的矩陣圖表的第一個axes
ax1 = fig.add_subplot(1, 3, 1)
#設定ax_1的標題
ax1.set_title(label="ax_1")
# 3.1*3的矩陣圖表的第二個axes
ax2 = fig.add_subplot(1, 3, 2)
#設定ax_2的標題
ax2.set_title(label="ax_2")
# 4.1*3的矩陣圖表的第三個axes
ax3 = fig.add_subplot(1, 3, 3)
#設定ax_3的標題
ax3.set_title(label="ax_3")
# 5.顯示圖
plt.show()
```
