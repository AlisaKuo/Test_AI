```python
cities = ['台北','台中','高雄','台南']
cities
```

```python
# 負向索引index
cities[-1] #最後一個
```

```python
# slice[前面的索引:後面的索引]
# 前面的索引包含,後面的索引不包含
cities[0:1]  #0
cities[0:2]  #0,1
cities[:2]   #不寫等於從最前面開始,就是0
cities[2:4]
cities[2:]   #後面不寫表示抓到最後一個
```
