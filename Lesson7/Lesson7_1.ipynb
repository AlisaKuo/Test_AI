```python
city1 = "台北"
city2 = "台中"
city3 = "高雄"
print (city1)
print (city2)
print (city3)
```

```python
#[]建立list實體
cities = ["台北", "台中","高雄"]    #[]建立一個list, 裡面放三筆字串, 又稱3個元素
print (cities)
print (cities[0])  #呼叫索引第0個
print (cities[1])
print (cities[2])
#print (cities[3]) --> index error, 因為沒有第3個元素
```
