```python
#讓使用者輸入直角三角形的對邊
#讓使用者輸入直角三角形的鄰邊
#計算斜邊
import math
side= int(input("請輸入對邊:"))
another_side= int(input("請輸入鄰邊:"))
result= math.sqrt(side**2+another_side**2)
print("斜邊是:", result)
```

```python
# 字串和數值變數的整合
fruits = 10
apple = 7
banana = 3

result="我有水果"+ str(fruits)+"個, 其中蘋果"+str(apple)+"顆, 香蕉"+str(banana)+"根"
print(result)
```

```python
# 字串和數值變數的整合
# 利用複合指令
fruits = 10
apple = 7
banana = 3

result = ""

result += "我有水果"
result += str(fruits)
result +="個,其中蘋果"
result += str(apple)
result += "顆,香蕉"
result += str(banana)+"根"
print(result)
```

```python
# 字串和數值變數的整合
# 格式化字串 (字串插補)
fruits = 10
apple = 7
banana = 3

result= f"我有水果{fruits}個,其中蘋果{apple}顆,香蕉{banana}根"
print(result)
```

```python
# 全部都在root區
# 流程控制的判斷
# 判斷讓執行的程式敘述不一樣 (變有智慧)

age= int(input("請輸入age:"))

if age >= 18:
    print("已滿18")
    print("可以考駕照了")
else:
    print("未滿18")
    print("不可以考駕照")
```
