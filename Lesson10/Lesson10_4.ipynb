```python
#自訂class
class Person:
    def __init__(self,name:str,age:int):  #自訂的初始化
        self.name = name    #建立name屬性
        self.age = age      #建立age屬性
    #實體方法
    def description(self):
        print(f"我的名字:{self.name}")
        print(f"我的年齡:{self.age}")
```

```python
p1=Person(name="Robert",age=40)
p1.description()              #呼叫實體方法
p2=Person(name="Ben",age=39)
p2.description()
```

```python
# Student 繼承 Person
# Person 是父類別
# Student 是子類別

class Student(Person):     #繼承
    @classmethod
    def nickname(cls):    #需要有參數的
        print("我是class的method")
    
    @staticmethod         #不需要有參數的
    def heroname():
        print("我是static的method")
        
    def __init__(self,name:str,age:int,chinese:int=60,english:int=60,math:int=60):
        super().__init__(name,age)
        #self.chinese = chinese #attribute
        self.__chinese = chinese  #＿＿變成引藏版, 只能讀不能改
        self.__english = english
        self.__math = math

    @property
    def chinese (self) -> int:
        return self.__chinese     #把隱藏版的值傳出去
    @property
    def english (self) -> int:
        return self.__english     #把隱藏版的值傳出去
    @property
    def math (self) -> int:
        return self.__math        #把隱藏版的值傳出去

    #複寫父類別(override)
    def description(self):
        super().description()
        print(f"chinese:{self.chinese}")
        print(f"english:{self.english}")
        print(f"math:{self.math}")
    


```

```python
s1 =Student(name="Alice",age=38,english=85,math=98)
#print (s1.chinese)
#print (s1.english)
#print (s1.math)
s1.description()
```
