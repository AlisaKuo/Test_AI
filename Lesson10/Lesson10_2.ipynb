```python
try:
    money= int(input("請輸入金額:"))
except:
    print("出現錯誤")
else:
    print(money)
finally:    
    print ("應用程式結束")
```

```python
try:
    money= int(input("請輸入金額:"))
    if money < 0:
        raise Exception("金額不可以小於0")
except ValueError:              #except 可以多個
    print("格式錯誤")
except Exception as error:      #Exception 是指所有錯誤, 把錯誤轉到變數 error
    print(error)
else:
    print(money)
finally:    
    print ("應用程式結束")
```
