```python
#自訂class
class Person:
    #def __init__(self):         #self不需要引數值來對應
    def __init__(self,name:str,age:int):  #引數位置呼叫
        self.name = name    #建立name屬性
        self.age = age      #建立age屬性
    
```

```python
p1 = Person("robert",30)
#print(type(p1))
print(p1.name)
print(p1.age)
```

```python
p2 = Person(age=35,name="alisa")    #不用按照順序
print(p2.name)
print(p2.age)
```
