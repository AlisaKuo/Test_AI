```python
#Support Chinese font in matplotlib
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")
```

```python

import numpy as np
import matplotlib.pyplot as plt

# 指定間隔數量,傳出固定間隔的數字
x = np.linspace(-3, 3, 50)

# 陣列運算
y1 = 2 * x + 1
y2 = x ** 2

plt.plot(x, y1, color='red', label='y=2x+1', linewidth=2, linestyle='--')
plt.plot(x, y2, label='y=x^2', linewidth=2, color='yellow', linestyle='-')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Two Functions')
plt.legend()
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.grid()    #畫格線
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

t=np.arange(0., 5., 0.2)
# 正弦波
y1 = np.sin(2 * np.pi * t)
# 指數衰減
y2 = np.exp(-t)
plt.plot(t, y1, 'bo--', label='sin')
plt.plot(t, y2, 'rs--', label='exp')
plt.xlabel('時間 (秒)')
plt.ylabel('振幅')
plt.title('阻尼正弦波')
plt.legend()
plt.grid()
plt.show()

```
